# All correllation that means something important

```python
for j in df.corr():
    for i in range(len(df.corr()[j])):
        if df.corr()[j][i] == 1 or df.corr()[j][i] == -1:
            continue
        if df.corr()[j][i] > 0.5 or df.corr()[j][i] < -0.5:
            print(df.corr().index[i], j, df.corr()[j][i])
```

As we see there no negative important correlations that statistifical means something between vars. Only positive correlational coefficents looking here.

# Correlation SalePrice between other
There we list most correlated vars with SalePrice

```python
for i in range(len(df.corr()['SalePrice'])):
    if df.corr()["SalePrice"][i] > 0.55:
        print(df.corr()["SalePrice"][i], df.corr()["SalePrice"].index[i])
```
